## indexing, slicing

`iloc` vs `loc`
*   `iloc` : 정수 인덱스 기반으로 원소를 참조.   list의 인덱스를 사용하는 방법과 동일.
*   `loc` : 축 레이블 기반으로 원소를 참조.

```python
s.iloc[0]   # Series의 첫번째 원소 참조(선택)
```

```python
s.iloc[-1]   # Series의 마지막 원소를 참조(선택).
```

```python
s.iloc[:2]  # slicing 의 시작 인덱스 0은 생략 가능. 첫번째 원소부터 2번째 원소까지 slicing
```

```python
s.iloc[-2:]  # 마지막 2개 원소를 slicing.  음수 인덱스 사용 가능.  마지막 인덱스 생략 가능
```

```python
s.loc[0]  #  Series에서 레이블이 0인 원소를 참조.
```

```python
s.loc[3]   # Series에서 레이블이 3인 원소를 참조
```

```python
s.loc[-1]  # KeyError, ValueError  - 리스트처럼 자동으로 부여되는 음수 인덱스는 없음,
```

## boolean indexing

*   `df[조건식]` 과 `df.loc[조건식]`은 같은 코드.
*   조건식을 만족하는 모든 행들을 선택.

```python
# 서울 데이터만 선택
df[df['city'] == '서울']
```

```python
# 2024년 데이터를 선택
df[df['year'] == 2024]  # df에서 선택하겠다 [df~~]이런 조건들을
```

Series, DataFrame에서 boolean indexing을 사용할 때 주의할 점:
*   and, or, not 키워드를 사용할 수 없음.
*   대신에 `&, |, ~` 연산자를 사용
*   반드시 (0를 사용해서 논리 연산의 순서를 명시해야 함.

```python
# 2021, 2022년 데이터 선택    # 반드시 괄호로 묶어야 함
df[(df['year'] == 2022) | (df['year'] == 2023)]
df[(df.year == 2022) | (df.year == 2023)]
```

```python
# 도시가 서울이거나 pop의 값이 1.2 이상 데이터 선택
df[(df['city'] == '서울') | (df['pop'] >= 1.2)]
# df[(df.city == '서울') | (df['pop'] >= 1.2)]
```
