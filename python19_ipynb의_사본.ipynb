생존자들 중 남성 생존자 비율,여성 -비율
*   남성 생존자 비율 = (남성 생존자 수)/(전체 생존자 수)
*   여성- =(여성-)/(전체-)

```python
# 전체 생존자
num_of_total_alive = len(tt[tt.alive == 'yes'])
num_of_total_alive
```

```python
print('남성 생존자 비율 = ', num_of_male_alive / num_of_total_alive)
print('여성 생존자 비율 = ', num_of_female_alive / num_of_total_alive)
```

생존자들 중 성인 생존자 비율, 어린이 생존자 비율
*   성인 생존자 비율 = (성인 생존자 수)/ (전체 생존자 수)
*   어린이- =(어린이-)/(전체-)


```python
# 성인 생존자 수
num_of_adult_alive = tt[(tt.who != 'child') & (tt.alive == 'yes')]
len(num_of_adult_alive)
```

```python
# 어린이 생존자 수
num_of_child_alive = tt[(tt.who == 'child') & (tt.alive == 'yes')]
len(num_of_child_alive)
```

```python
print('성인 생존자 비율 = ', len(num_of_adult_alive) / num_of_total_alive  )
print('어린이 생존자 비율 = ', len(num_of_child_alive) / num_of_total_alive  )
```

## 객실 등급별 생존자 비율

*1등급 객실 생존자 비율 = (1등급 객실 생존자)/ 전체 -
2
3

```python
# 1등급 객실 생존자
num_of_first_alive = len(tt[(tt.pclass == 1) & (tt.alive == 'yes')])
num_of_first_alive
```

```python
# 2등급 객실 생존자
num_of_second_alive = len(tt[(tt.pclass == 2) & (tt.alive == 'yes')])
num_of_second_alive
```

```python
# 3등급 객실 생존자
num_of_third_alive = len(tt[(tt.pclass == 3) & (tt.alive == 'yes')])
num_of_third_alive
```

```python
print('1등급 객실 생존자 비율 = ', num_of_first_alive / num_of_total_alive )
print('2등급 객실 생존자 비율 = ', num_of_second_alive / num_of_total_alive )
print('3등급 객실 생존자 비율 = ', num_of_third_alive / num_of_total_alive )
```

```python
tt.pclass.unique()
```

```python
for c in tt.pclass.unique():
    num_of_class_alive = len(tt[(tt.pclass == c) & (tt.alive == 'yes')])
    print(f'{c}등급 객실 생존자 비율 = {num_of_class_alive/ num_of_total_alive}')
```

* 1등급 객실 생존자 비율 = (1등급 객실 생존자 수) / (1등급 객실 승객수)
* 2
* 3

```python
for c in tt.pclass.unique():
    # 각 객실 등급의 승객수
    num_of_class = len(tt[tt.pclass == c])
    # 각 객실의 등급의 생존자 수
    num_of_class_alive = len(tt[(tt.pclass == c) & (tt.alive == 'yes')])
    # 각 객실 등급의 생존자 비율
    print(f'{c}등급 객실에서 생존자 비율= , {num_of_class_alive / num_of_class}')
```
