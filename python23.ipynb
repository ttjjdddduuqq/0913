# DataFrame Merge(Join)

```python
emp = pd.DataFrame(data={
    'empno':[100, 101, 200, 201],
    'ename': ['홍길동', '오쌤', 'Scott', 'King'],
    'deptno': [10, 10, 20, 40]
})
emp
```

```python
dept = pd.DataFrame(data={
    'deptno': [10,20,30],
    'dname': ['IT', 'HR', ' Sales']
})
dept
```

*   `pd.merge(df1, df2, how, on, ...)` 함수(function)
*   `pd.DataFrame,merge(df, how, on, ..)` 메섣,(method)
*   `how` 파라미터: merge(join) 방식 설정.
    *   inner, left, right, outer
    *   merge의 기본값은 inner. how = 'inner'(기본값은 생략 가능)
*   `on` 파라미터: merge(join)의 기준이 되는 컬럼(컬럼들의 리스트)를 설정.
    *   merge(join)를 하려는 2개의 데잍프레임에서 컬럼 이름이 같은 경우에는  on 파라미터 설정을 생략 가능.

## inner join
 (두 데이터의 교집합을 나타냄)

```python
pd.merge(emp, dept)
# how = 'inner', on='deptno' 생략
```

```python
emp.merge(dept)
```

```python
dept.merge(emp) #먼저 쓰면 (dept) 먼저 나열 괄호()안에(emp)는 뒤에 나옴
```

## left (outer) join
왼쪽데이터 + 교집합

```python
pd.merge(emp, dept, how='left')
```

## right (outer) join
오른쪽 데이터 + 교집합

```python
pd.merge(emp, dept, how='right')
```

## (full) outer join
두 데이터의 전체 데이터

```python
pd.merge(emp, dept, how='outer')
```

# DataFrame 모양 바꾸기(Reshape)

## `melt`

`pd.DataFrame.melt()` 메서드: wide-form 데이터를 long-form 데이터로 변환

*   `id_vars` 파라미터: melting 동작을 할 때 컬럼으로 유지할 컬럼 이름.
    *   `id_vars` 로 설정하지 않은 모든 변수(컬럼)들은 `variable` 컬럼으로 melting 됨.
    *    `id_vars` 로 설정하지 않은 모든 변수(컬럼)들의 셀에 있는 값들은 `value` 컬럼으로 melting됨.
*   `var_name` 파라미터 : `variable` 컬럼의 이름으로 사용할 문자열.
*   `value_name` 파라미터 : `value` 컬럼의 이름으로 사용할 문자열.

```python
df = pd.DataFrame({
    'gender': ['Female', 'Male'],
    'lunch': [10, 5],
    'dinner' : [20, 30]
})
df
```

```python
df.melt(id_vars=['gender'])
# lunch 와 dinner 라는 컬럼이 변수 컬럼으로 바뀜
# 10 , 20 , 5 , 30 은 value라는 값으로 바뀜
```

```python
df.melt(id_vars=['gender'], var_name = 'time', value_name='size')
```

## `pivot`

`pd.DataFrame.pivot()` 메서드 :long-form(세로로 긴) 데이터프레임을 wide-form(가로로 긴) 데이터 프레임으로 변환.

*   `index` 파라미터: pivoting한 데이터프레임 row index로 사용할 컬럼(들)의 이름.
*   `columns` 파라미터 : pivoting한 데이터프레임에서 컬럼 이름으로 사용할 컬럼(들)의 이름
*   `values` 파라미터: pivoting한 데이터프레임의 각 셀에 채울 값들이 있는 컬럼(들)의 이름


```python
df2 = pd.DataFrame({
    'time' : ['Lunch'] * 3 + ['Dinner'] * 3,
    'day' : ['Fri', 'Sat', 'Sun'] * 2,
    'total_bill': np.arange(10, 70, 10),
    'tip': np.arange(1,7)
})
df2
```

```python
df2.pivot(index='time', columns='day',values='tip')
```

```python
result = df2.pivot(index='time', columns='day', values=['total_bill','tip'])
result
```

```python
result.columns
```

```python
df2.pivot(index='day', columns='time', values='tip')
```

```python
df2.pivot(index='day', columns='time', values=['total_bill', 'tip'])
```

# groupby vs pivot_table

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## groupby

```python
# 성별 팁의 평균
tips.groupby(by=['sex'], observed=True).tip.mean()
```

```python
# 성별 , 흡연 여부 팁의 평균
tips.groupby(by=['sex', 'smoker'], observed=True).tip.mean()
```

```python
# 성별 흡연 여부별 팁/영수증 평균
result = tips.groupby(by=['sex', 'smoker'],observed=True)[['tip', 'total_bill']].mean()
result
```

```python
result = result.reset_index()
result
```

```python
result.pivot(index='sex', columns='smoker', values=['tip', 'total_bill'])
```

## pivot table

`pd.DataFrame.pivot_table()` 메서드 파라미터:
*   `values` :  pivoting을 해서 각 셀에 값으로 설정할 변수(컬럼)(들).
*    `index` : pivoting을 할 때 row index로 설정할 변수(컬럼)(들).
*   `columns` : pivoting을 할 때 컬럼 이름으로 설정할 변수(컬럼)(들).
*   `aggfunc` : aggregation function(집계 함수) . 통계 값을 계산하는 함수 이름 또는 함수 계체.
    *   aggfunc 파라미터의 기본값은 'mean'.

```python
# 성별 팁의 평균
tips.pivot_table(values='tip', index='sex',observed=True) #  aggfunc='mean' 생략
```

```python
# 성별 흡연 여부별 팁의 평균
tips.pivot_table(values='tip', index=['sex','smoker'], observed=True)
```

```python
tips.pivot_table(values='tip', index='sex', columns='smoker', observed=True)
# 인덱스 , 컬럼 = 범주 형태를 입력해야
```

```python
# 성별 흡연여부별 팁/영수증 평균
tips.pivot_table(values=['tip', 'total_bill'], index=['sex', 'smoker'],observed=True )
```

```python
tips.pivot_table(values=['tip', 'total_bill'], index='sex', columns='smoker', observed=True)
```

```python
tips.pivot_table(values=['tip', 'total_bill'], index='smoker', columns='sex', observed=True)
```
